```
req = urllib.request.urlopen('https://upload.wikimedia.org/wikipedia/en/thumb/4/41/Flag_of_India.svg/1200px-Flag_of_India.svg.png')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img = cv2.imdecode(arr, -1)
I = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(I)
print(I.shape)
print(I.shape[0]*I.shape[1])
```

```
U, s, V = np.linalg.svd(I, full_matrices=True)
print(U.shape,s.shape,V.shape)
S = np.zeros((800, 1200))
S[:800,:800] = np.diag(s)
x = np.random.randint(16, size=(5)).astype('uint8')
comp = np.zeros((800, 1200))
b = np.dot(U,np.dot(S,V))
cv2_imshow(b)
```
