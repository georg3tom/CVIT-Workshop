```
def distance_from_origin(a,b,c):
  import math
  return abs(c)/math.sqrt(a*a+b*b)
distance_from_origin(2,2,1)
```

```
#tanh
from math import e
X = np.linspace(start = -5,stop =5,num =100)
Y = np.tanh(X)
%matplotlib inline
plt.plot(X, Y)
```

```
#gaussian
X = np.linspace(start = -5,stop =5,num =100)
Y = np.exp(-X**2)
%matplotlib inline
plt.plot(X, Y)
```

```
#step
X = np.linspace(start = -5,stop =5,num =100)
Y = np.array([ 1 if x >= 0 else 0 for x in X])
%matplotlib inline
plt.plot(X, Y)
```

```
#Leaky
X = np.linspace(start = -5,stop =5,num =100)
Y = np.array([ max(0.1*x,x) for x in X])
%matplotlib inline
plt.plot(X, Y)
```

```
%matplotlib inline
x = np.linspace(-2, 2, 200)
y = np.linspace(-2, 2, 200)
X, Y = np.meshgrid(x, y)
f = np.exp(-X*X - Y*Y)
C = plt.contour(X, Y, f)c
plt.clabel(C)
```

```
a = 1
b = 1
c = 0
d = 10
xs = np.linspace(-20,20,200)
zs = np.linspace(-20,20,200)
X,Z  = np.meshgrid(xs,zs)
Y = (-X*a-Z*c-d)*1.0/b
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(X,Y,Z)
```

```
from sklearn.datasets import make_swiss_roll
%matplotlib inline
X, _ = make_swiss_roll(1000)

xs = X[:, 0]
ys = X[:, 1]
zs = X[:, 2]
X, Y = np.meshgrid(xs, ys)
Z = np.exp(-X**2-Y**2)
plt3d = plt.figure().gca(projection='3d')
# Write code here to plot the 3D scatter-plot(~1 line)
plt3d.plot_surface(X,Y,Z)
```
